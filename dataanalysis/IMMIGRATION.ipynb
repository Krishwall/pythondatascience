## IMMIGRATION DATA ANALYSIS

```python
df=pd.read_excel('Canada.xlsx',sheet_name=1,skiprows=20,skipfooter=2)

```

```python
df=df.drop(columns=['REG','AREA'])
```

```python
df=df.rename(columns={'OdName':'Country',
                  'AreaName':'CONTINENT',
                  'RegName':'Region',
                  'DevName':'Status'})
df.head()
```

```python
df=df.drop(columns=['DEV','Type','Coverage'])

```

make all columns to lower case- optional

```python
df.columns=[str(name).lower() for name in df.columns.tolist() ]
df.head()
```

```python
df=df.set_index('country')
df.head()
```

```python
years=list(map(str,range(1980,2014)))
print(years)
```

```python
df[years].sum(axis=1)  # AXIS 1 for row wise sum ,default=0 column wise
```

```python
df['total']=df[years].sum(axis=1)
df.head()
```

```python
# top 25 countries with highest immigration
colors = ['red' if total > 500000 else 'blue' for total in df.total.sort_values(ascending=False)]
ax=df.sort_values('total',ascending=False).head(25).plot(
    kind='bar',
    y='total',
    figsize=(15,5),
    title='TOP 25 COUNTRIES WITH highest immigration',
        color=colors,
    alpha=1,
    ylabel='TOTAL',

    
)

ax.set_xlabel("COUNTRIES",fontsize=15)
```

```python


# Sort the DataFrame by 'total' in descending order and select top 25
df_sorted = df.sort_values('total', ascending=False).head(25)

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(20, 15))

# Define colors based on the condition (total > 500000)
colors = ['red' if total > 500000 else 'blue' for total in df_sorted['total']]

# Create the bar plot with customized colors
bars = ax.bar(df_sorted.index, df_sorted['total'], color=colors)

# Add labels and title
ax.set_xlabel("COUNTRIES", fontsize=15)
ax.set_ylabel("TOTAL", fontsize=15)
ax.set_title('TOP 25 COUNTRIES WITH highest immigration', fontsize=15)

# Adding a legend for clarity
red_patch = plt.Line2D([0], [0], color='red', lw=4, label='Total > 500,000')
blue_patch = plt.Line2D([0], [0], color='blue', lw=4, label='Total <= 500,000')
ax.legend(handles=[red_patch, blue_patch], fontsize=12)

# Function to annotate bars with values
def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
ax.set_xticklabels(df_sorted.index, rotation=90,fontsize=15,fontweight='bold')
# Call the function to add annotations
autolabel(bars)

plt.tight_layout()
plt.show()
```
