```python
from dputils.scrape import Scraper,Tag
```

```python
url=r"https://www.amazon.in/Apple-iPhone-15-128-GB/dp/B0CHX2F5QT"
s1=Scraper(url)
```

```python
heading=Tag('h1',cls="a-size-large a-spacing-none")
price=Tag('span',cls="a-price-whole")  # span is a tag

s1.get_data_from_page(heading=heading,price=price)
```

```python
i=1
watches=[]
while i<10:
    url=f'https://www.watchesofswitzerland.com/c/Watches/Mens-Watches?q=&page={i}&sort='

    s2=Scraper(url)

    results=s2.get_repeating_date_from_page(
        target=Tag('body'),
        
        brand=Tag(cls="productTileBrand"),
        productname=Tag(cls="productTileName"),
        price=Tag(cls="productTilePrice")
    
    )
    watches.extend(results)
    i+=1
watches
```

```python
watchdf=pd.DataFrame(watches).drop_duplicates().dropna()
watchdf
```
